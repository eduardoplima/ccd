# Automação dos processos de Nereu

```python
import os
import sys
import pymssql
import locale

sys.path.append(os.path.abspath(".."))

import pandas as pd
import numpy as np


from dotenv import load_dotenv
from num2words import num2words
from docxtpl import DocxTemplate
from docx2pdf import convert
from utils_ccd import get_connection, generate_pdf_office

load_dotenv()
conn = get_connection()
locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8')
```

```python
nereu_sql = f''' 
SELECT 
    CONCAT(pro.numero_processo, '/', pro.ano_processo) as processo,
    pro.assunto,
    rel.nome as relator
    
FROM processo.dbo.Processos pro
    INNER JOIN processo.dbo.Relator rel ON pro.codigo_relator = rel.codigo
WHERE 
    pro.setor_atual = 'CCD'
    AND CONCAT(pro.numero_processo, '/', pro.ano_processo) IN (
  '002161/2025', '002160/2025', '002156/2025', '001860/2025', '001858/2025', 
  '001856/2025', '001854/2025', '001853/2025', '001852/2025', '001847/2025', 
  '001846/2025', '001845/2025', '001844/2025', '000873/2025', '000855/2025', 
  '002018/2025', '003691/2022', '003673/2022', '003070/2022', '000124/2022'
)   
'''
processos_nereu = pd.read_sql(nereu_sql, conn)
```

```python
for row in processos_nereu.itertuples():
    doc = DocxTemplate("templates/info_nereu.docx")
    
    context = {
        'processo': row.processo,
        'relator': row.relator
    }
    processo_underline = row.processo.replace('/', '_')

    doc.render(context)
    doc.save(f"../../saidas/automacao/nereu/{processo_underline}.docx")
    generate_pdf_office(f"../../saidas/automacao/nereu/{processo_underline}.docx", f"../../saidas/automacao/nereu/")
    print(f"Gerado arquivo para o processo {row.processo}")
```
