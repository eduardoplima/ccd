```python
def get_file_path(row):
    return Path(dir_informacoes) / row['setor'].strip() / row['arquivo']
```

```python
df_informacoes['caminho_arquivo'] = df_informacoes.apply(get_file_path, axis=1)
```

```python
df_informacoes['processo'] = df_informacoes['numero_processo'] + '/' + df_informacoes['ano_processo']
df_informacoes['dir_processo'] = df_informacoes['numero_processo'] + '_' + df_informacoes['ano_processo']
```

```python
for processo in processos:
    df = df_informacoes[df_informacoes['processo'] == processo]
    dir_processo = processo.replace('/', '_')
    arquivos = df['caminho_arquivo'].values
    print(f'Processo {processo} tem {len(arquivos)} arquivos')
    if not os.path.exists(f'arquivos_processos/{dir_processo}'):
        os.mkdir(f'/home/eduardo/Dev/ccd/scripts/arquivos_processos/{dir_processo}')
    for arquivo in arquivos:
        print(f'Copiando {arquivo} para arquivos_processos/{dir_processo}')
        try:
            shutil.copyfile(arquivo, f'arquivos_processos/{dir_processo}/{arquivo.name}')
        except FileNotFoundError as e:
            print(f'Arquivo não encontrado: {arquivo}')
            print(e)
        except Exception as e:
            print(f'Erro ao copiar arquivo {arquivo}')
```
