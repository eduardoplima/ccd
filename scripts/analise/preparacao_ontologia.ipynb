```python
import re

def extrair_conclusao(texto):
    """
    Extrai o bloco iniciado exatamente por uma quebra de linha
    seguida por 'CONCLUSÃO' em maiúsculas.
    """
    # \n\s*CONCLUSÃO  → deve estar em nova linha (admite espaços após a quebra)
    padrao = r'\n\s*CONCLUSÃO\s*(.*)$'

    match = re.search(padrao, texto, flags=re.DOTALL)
    if match:
        # match.group(1) = somente o texto depois de "CONCLUSÃO"
        return "CONCLUSÃO\n" + match.group(1).strip()
    return None
```

```python
df_obrigacoes['processo'] = df_obrigacoes['numero_processo'].astype(str).str.zfill(6) + '/' + df_obrigacoes['ano_processo'].astype(str)
```

```python
small_obrigacoes = df_obrigacoes[:100]
```

```python
infos_obrigacoes = {}
```

```python
for idx, row in small_obrigacoes.iterrows():
    print(f"Processando {idx+1}/{len(small_obrigacoes)}: {row['processo']}")
    processo = row['processo']
    infos = get_informacoes_processo(processo, conn)
    infos_obrigacoes[processo] = infos
```

```python
processos = set(small_obrigacoes['processo'].values)
```

```python
processos = processos - set(infos_obrigacoes.keys())
```

```python
for processo in processos:
    print(f"Processando {processo}")
    infos = get_informacoes_processo(processo, conn)
    infos_obrigacoes[processo] = infos
```

```python
len(infos_obrigacoes.keys())
```

```python
conclusoes = []
for key in infos_obrigacoes.keys():
    voto = infos_obrigacoes[key][infos_obrigacoes[key]['resumo'].str.contains('Voto')]
    if not voto.empty:
        texto_voto = voto['texto'].values[0]
        conclusao = extrair_conclusao(texto_voto)
        if conclusao:
            conclusoes.append(conclusao)
```
